The model is defined with the sequential API with 10 neurons and the weight is initialized to zeros with the softmax activation function. The ANN is only one layer and expecting an input with shape of 784 * 1

```python
epochs = 200
batch_size = 128
verbose = 1
nb_classes = 10
n_hidden = 128
validation_split = 0.2
reshaped = 784

mnist = keras.datasets.mnist

(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

# reshaping the images
X_train = X_train.reshape(60000, reshaped)
X_test = X_test.reshape(10000, reshaped)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# normalize the inputs to be within [0, 1]
print(f"Before normalizing the X_train: {X_train[:10]}")
X_train /= 255
X_test /= 255
print(f"After normalizing the X_train: {X_train[:10]}")

print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')

# one hot representation of the labels
print(f"Before one hot encoding Y_train: {Y_train[:10]}")
Y_train = tf.keras.utils.to_categorical(Y_train, nb_classes)
Y_test = tf.keras.utils.to_categorical(Y_test, nb_classes)
print(f"After one hot encoding Y_train: {Y_train[: 10]}")
```

```python
# build the model
model = tf.keras.models.Sequential()
model.add(keras.layers.Dense(nb_classes, input_shape=(reshaped, ), name = "dense_layer_1", activation = keras.activations.softmax))
```

```python
# compile the model

model.compile(optimizer="SGD", loss="categorical_crossentropy", metrics=['accuracy'])
```

```python
# training the model
# epochs is the number of times the model is exposed to the training set
# batch_size is the number of  training instances observed before the optimizer updates the weight
model.fit(X_train, Y_train, batch_size=batch_size, epochs=epochs, verbose=verbose, validation_split=validation_split)
```

```python
score = model.evaluate(X_test, Y_test)
print("score: {}".format(score[1]))
```

```python
import tensorflow as tf
from tensorflow import keras

#improving the accuracy by adding more layers
epochs = 50
batch_size = 128
verbose = 1
nb_classes = 10
n_hidden = 128
validation_split = 0.2

# loading the mnist data
mnist = keras.datasets.mnist
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

reshaped = 784

X_train = X_train.reshape(60000, reshaped)
X_test = X_test.reshape(10000, reshaped)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

#normalizing the data to be [0,1]
X_train /= 255
X_test /= 255

# one hot encoding the values
Y_train = keras.utils.to_categorical(Y_train, nb_classes)
Y_test = keras.utils.to_categorical(Y_test, nb_classes)


#build the model
model = keras.models.Sequential()
model.add(keras.layers.Dense(n_hidden, input_shape=(reshaped,), name="dense_layer_1", activation=keras.activations.relu))
model.add(keras.layers.Dense(n_hidden, name= "dense_layer_2", activation="relu"))
model.add(keras.layers.Dense(nb_classes, name="dense_layer_3", activation="softmax"))

# summary
model.summary()

# compiling the model
model.compile(optimizer="SGD", loss="categorical_crossentropy", metrics=['accuracy'])

# training the model
model.fit(X_train, Y_train, epochs=epochs, batch_size=batch_size, verbose=verbose, validation_split=validation_split)


#evaluating the model
score = model.evaluate(X_test, Y_test)
print("accuracy: ", score[1])
```

```python
import tensorflow as tf
from tensorflow import keras

#improving the accuracy by adding more layers
epochs = 50
batch_size = 128
verbose = 1
nb_classes = 10
n_hidden = 128
validation_split = 0.2
dropout = 0.3  # dropping out some the features for the model to generalized very well

# loading the mnist data
mnist = keras.datasets.mnist
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

reshaped = 784

X_train = X_train.reshape(60000, reshaped)
X_test = X_test.reshape(10000, reshaped)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

#normalizing the data to be [0,1]
X_train /= 255
X_test /= 255

# one hot encoding the values
Y_train = keras.utils.to_categorical(Y_train, nb_classes)
Y_test = keras.utils.to_categorical(Y_test, nb_classes)


#build the model
model = keras.models.Sequential()
model.add(keras.layers.Dense(n_hidden, input_shape=(reshaped,), name="dense_layer_1", activation=keras.activations.relu))
model.add(keras.layers.Dropout(dropout))
model.add(keras.layers.Dense(n_hidden, name= "dense_layer_2", activation="relu"))
model.add(keras.layers.Dropout(dropout))
model.add(keras.layers.Dense(nb_classes, name="dense_layer_3", activation="softmax"))

# summary
model.summary()

# compiling the model
model.compile(optimizer="SGD", loss="categorical_crossentropy", metrics=['accuracy'])

# training the model
model.fit(X_train, Y_train, epochs=epochs, batch_size=batch_size, verbose=verbose, validation_split=validation_split)


#evaluating the model
score = model.evaluate(X_test, Y_test)
print("accuracy: ", score[1])
```

Adding the dropout did not improve the previous score of the model. Therefore , testing with a different optimizer.
