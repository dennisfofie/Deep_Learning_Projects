```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d
from sklearn.datasets import make_blobs

# Generate example data
np.random.seed(0)
data, _ = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=0.1)

# Use KMeans to cluster the data and obtain centroids
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(data)
centroids = kmeans.cluster_centers_

# Compute Voronoi diagram around centroids
vor = Voronoi(centroids)

# Plot Voronoi cells
voronoi_plot_2d(vor)

# Plot data points
plt.scatter(data[:, 0], data[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.7)

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red', label='Centroids')

plt.title('Voronoi Diagram with K-Means Clustering')
plt.xlabel('Distance')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate example data
np.random.seed(0)
data, _ = make_blobs(n_samples=300, centers=10, random_state=42, cluster_std=0.1)

# Use KMeans to cluster the data and obtain centroids
kmeans = KMeans(n_clusters=10, random_state=42)
kmeans.fit(data)
centroids = kmeans.cluster_centers_

# Compute Voronoi diagram around centroids
vor = Voronoi(centroids)

# Plot Voronoi cells
voronoi_plot_2d(vor)

# Plot data points
plt.scatter(data[:, 0], data[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.7)

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red', label='Centroids')

# Annotate centroids with their coordinates
for i, centroid in enumerate(centroids):
    plt.text(centroid[0], centroid[1], "{}, {}".format(centroid[0], centroid[1]), 
             fontsize=8, ha='left', va='bottom', color='black')

plt.title('Voronoi Diagram with K-Means Clustering')
plt.xlabel('Distance')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate example data
np.random.seed(0)
data, _ = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=0.1)

# Use KMeans to cluster the data and obtain centroids
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(data)
centroids = kmeans.cluster_centers_

# Choose a centroid (e.g., the first one, index 0)
chosen_centroid = centroids[0]

# Get the indices of data points assigned to the chosen centroid
points_in_chosen_centroid = np.where(kmeans.labels_ == kmeans.predict([chosen_centroid]))[0]

# Plot all data points
plt.scatter(data[:, 0], data[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.7, label='All Data Points')

# Highlight data points assigned to the chosen centroid
plt.scatter(data[points_in_chosen_centroid, 0], data[points_in_chosen_centroid, 1], 
            marker='o', s=100, color='red', label='Chosen Centroid Points')

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red', label='Centroids')

plt.title('Data Points Grouped to a Chosen Centroid')
plt.xlabel('Distance')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.cluster import KMeans
import numpy as np


np.random.seed(0)
data, _ = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=0.1)


param_grid = {
    'n_clusters': [i for i in range(2, 50+1)],
    'init': ["k-means++", "random"],
    "max_iter": [100, 200, 300, 400, 500]
}

grid_search = GridSearchCV(KMeans(random_state=42), param_grid=param_grid, cv=10, scoring=['accuracy'])
grid_search.fit(data)
best_kmeans = grid_search.best_estimator_
print(best_kmeans)
```
